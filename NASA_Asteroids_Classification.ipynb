```python
import warnings
# Suppress FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Function to randomly introduce null values
def introduce_null_values(df, fraction=0.1):
    """
    Randomly introduces null values into the dataset.
    Args:
        df: DataFrame to modify.
        fraction: Fraction of values to replace with nulls (default 10%).
    Returns:
        DataFrame with introduced null values.
    """
    df_copy = df.copy()
    total_values = df_copy.size
    n_nulls = int(total_values * fraction)

    for _ in range(n_nulls):
        i = np.random.randint(0, df_copy.shape[0])
        j = np.random.randint(0, df_copy.shape[1])
        col_name = df_copy.columns[j]
        col_dtype = df_copy.dtypes[col_name]

        try:
            if pd.api.types.is_numeric_dtype(col_dtype):
                df_copy.iat[i, j] = np.nan
            elif pd.api.types.is_object_dtype(col_dtype):
                df_copy.iat[i, j] = np.nan
            elif pd.api.types.is_bool_dtype(col_dtype):
                df_copy[col_name] = df_copy[col_name].astype(object)  # Convert bool to object
                df_copy.iat[i, j] = np.nan
            else:
                print(f"Skipping column '{col_name}' with dtype '{col_dtype}'")
        except Exception as e:
            print(f"Error introducing NaN in column '{col_name}': {e}")
    return df_copy

# Function to introduce typographical errors
def introduce_typo_errors(df, fraction=0.1):
    """
    Introduces typographical errors in string columns.
    Args:
        df: DataFrame to modify.
        fraction: Fraction of values in each string column to introduce errors (default 10%).
    Returns:
        DataFrame with introduced typographical errors.
    """
    df_copy = df.copy()
    string_columns = df_copy.select_dtypes(include=['object']).columns

    for col in string_columns:
        n_errors = int(df_copy[col].notnull().sum() * fraction)
        indices = np.random.choice(df_copy[df_copy[col].notnull()].index, n_errors, replace=False)
        for idx in indices:
            value = df_copy.at[idx, col]
            try:
                if isinstance(value, str) and len(value) > 2:
                    pos = np.random.randint(0, len(value))
                    typo_char = chr(np.random.randint(97, 123))
                    df_copy.at[idx, col] = value[:pos] + typo_char + value[pos + 1:]
            except Exception as e:
                print(f"Error introducing typo in column '{col}' at index {idx}: {e}")
    return df_copy

# Function to add noise to numerical values
def add_noise_to_numerical(df, std_dev=0.1):
    """
    Adds random noise to numerical columns.
    Args:
        df: DataFrame to modify.
        std_dev: Standard deviation of the noise (default 10% of each value).
    Returns:
        DataFrame with added noise.
    """
    df_copy = df.copy()
    numeric_columns = df_copy.select_dtypes(include=[np.number]).columns

    for col in numeric_columns:
        try:
            if df_copy[col].std() > 0:  # Ensure column has variability
                noise = np.random.normal(0, std_dev * df_copy[col].std(), size=df_copy[col].shape)
                df_copy[col] += noise
        except Exception as e:
            print(f"Error adding noise to column '{col}': {e}")
    return df_copy

# Function to shuffle values within a column
def shuffle_columns(df):
    """
    Randomly shuffles the values within all columns.
    Args:
        df: DataFrame to modify.
    Returns:
        DataFrame with shuffled column values.
    """
    df_copy = df.copy()
    for column in df_copy.columns:
        try:
            df_copy[column] = np.random.permutation(df_copy[column].values)
        except Exception as e:
            print(f"Error shuffling column '{column}': {e}")
    return df_copy

# Automatically corrupt the dataset
def corrupt_dataset(df, null_fraction=0.1, typo_fraction=0.05, noise_std_dev=0.1):
    """
    Applies all corruption techniques to the dataset.
    Args:
        df: DataFrame to modify.
        null_fraction: Fraction of values to replace with nulls.
        typo_fraction: Fraction of string values to introduce typos.
        noise_std_dev: Standard deviation of noise for numerical values.
    Returns:
        Corrupted DataFrame.
    """
    df = introduce_null_values(df, fraction=null_fraction)
    df = introduce_typo_errors(df, fraction=typo_fraction)
    df = add_noise_to_numerical(df, std_dev=noise_std_dev)
    df = shuffle_columns(df)
    return df

# Corrupt the dataset
corrupted_df = corrupt_dataset(df, null_fraction=0.1, typo_fraction=0.05, noise_std_dev=0.1)
```

```python
# Function to validate if null values, typos, noise, and shuffling were correctly applied
def validate_corruption(original_df, corrupted_df):
    """
    Validates the corruption applied to the dataset by comparing the original and corrupted dataframes.
    Args:
        original_df: Original DataFrame before corruption.
        corrupted_df: Corrupted DataFrame after corruption.
    Returns:
        Dictionary with validation results for each corruption type.
    """
    validation_results = {
        "Null Values Introduced": (corrupted_df.isna().sum() > original_df.isna().sum()).any(),
        "Data Shuffling Detected": False,
        "Typographical Errors Detected": False,
        "Noise Added to Numerical Columns": False,
    }

    # Check for shuffling
    for col in original_df.columns:
        if not original_df[col].equals(corrupted_df[col]):
            validation_results["Data Shuffling Detected"] = True
            break

    # Check for typographical errors in string columns
    string_columns = original_df.select_dtypes(include=["object"]).columns
    for col in string_columns:
        if not original_df[col].equals(corrupted_df[col]):
            validation_results["Typographical Errors Detected"] = True
            break

    # Check for noise in numerical columns
    numeric_columns = original_df.select_dtypes(include=[np.number]).columns
    for col in numeric_columns:
        if not (original_df[col] == corrupted_df[col]).all():
            validation_results["Noise Added to Numerical Columns"] = True
            break

    return validation_results


# Run validation
validation_results = validate_corruption(df, corrupted_df)
print(validation_results)
```

```python
# Function to handle missing values
def handle_missing_values(df):
    """
    Handles missing values in the dataset by:
    - Dropping columns with more than 50% missing values.
    - Filling numerical columns with the mean.
    - Filling categorical columns with the mode.
    """
    # Drop columns with more than 50% missing values
    threshold = len(df) * 0.5
    df = df.dropna(thresh=threshold, axis=1)

    # Fill missing values for numerical columns with the mean
    numeric_columns = df.select_dtypes(include=[np.number]).columns
    for col in numeric_columns:
        df[col] = df[col].fillna(df[col].mean())

    # Fill missing values for categorical columns with the mode
    categorical_columns = df.select_dtypes(include=['object']).columns
    for col in categorical_columns:
        if not df[col].isnull().all():  # Avoid issues with completely empty columns
            df[col] = df[col].fillna(df[col].mode()[0])

    return df

# Function to correct typographical errors
def correct_typo_errors(df):
    """
    Corrects typographical errors in string columns by:
    - Standardizing text to lowercase.
    - Stripping leading and trailing spaces.
    """
    string_columns = df.select_dtypes(include=['object']).columns
    for col in string_columns:
        # Standardize text
        df[col] = df[col].str.lower().str.strip()

    return df

# Function to clean numerical values
def clean_numerical_values(df):
    """
    Cleans numerical columns by:
    - Removing outliers based on the IQR method.
    """
    numeric_columns = df.select_dtypes(include=[np.number]).columns
    for col in numeric_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

    return df

# Main function to clean the dataset
def clean_dataset(df):
    """
    Cleans the entire dataset by applying all cleaning functions.
    """
    df = handle_missing_values(df)
    df = correct_typo_errors(df)
    df = clean_numerical_values(df)
    return df

# Clean the corrupted dataset
cleaned_df = clean_dataset(corrupted_df)
```

```python
from sklearn.preprocessing import LabelEncoder

# Load the cleaned dataset
df = cleaned_df.copy()

# Automatically detect column types
categorical_columns = df.select_dtypes(include=['object']).columns
date_columns = [col for col in categorical_columns if 'date' in col.lower()]
binary_columns = [col for col in df.columns if df[col].nunique() == 2 and df[col].dtype == 'object']
remaining_categorical = list(set(categorical_columns) - set(date_columns) - set(binary_columns))

# Ensure 'Hazardous' is encoded if it contains True/False values
if 'Hazardous' in df.columns:
    if df['Hazardous'].dtype == 'bool' or set(df['Hazardous'].unique()) <= {True, False}:
        df['Hazardous'] = df['Hazardous'].astype(int)

# Process date columns
for column in date_columns:
    df[column] = pd.to_datetime(df[column], errors='coerce')  # Convert to datetime
    df[f'{column}_year'] = df[column].dt.year
    df[f'{column}_month'] = df[column].dt.month
    df[f'{column}_day'] = df[column].dt.day
    df.drop(column, axis=1, inplace=True)  # Drop original date column

# Encode binary columns
label_encoder = LabelEncoder()
for column in binary_columns:
    df[column] = label_encoder.fit_transform(df[column].astype(str))

# Apply One-Hot Encoding for remaining categorical columns
df = pd.get_dummies(df, columns=remaining_categorical, drop_first=True)
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score, roc_auc_score,
    confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Ensure reproducibility
np.random.seed(42)

# Apply standard scaling to the feature matrix
scaler = StandardScaler()
STD_scaled_df = scaler.fit_transform(X)  # Scaled feature matrix

# Split the scaled dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(STD_scaled_df, y, test_size=0.2, random_state=42)

# Display the shapes of the resulting datasets
print("Shapes of scaled train and test sets:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
# Hyperparameter grids
param_grids = {
    "Random Forest": {
        "n_estimators": [50, 100, 200],
        "max_depth": [None, 10, 20],
        "criterion": ["gini", "entropy"],
    },
    "Gradient Boosting Machine": {
        "n_estimators": [50, 100, 200],
        "learning_rate": [0.01, 0.1, 0.2],
        "max_depth": [3, 5, 10],
    },
    "Decision Tree": {
        "criterion": ["gini", "entropy"],
        "max_depth": [None, 10, 20],
    },
    "AdaBoost": {
        "n_estimators": [50, 100, 200],
        "learning_rate": [0.01, 0.1, 0.5],
    },
    "K-Nearest Neighbors": {
        "n_neighbors": [3, 5, 7],
        "weights": ["uniform", "distance"],
    },
    "Gaussian Naive Bayes": {},  # No hyperparameters to tune
    "Linear Discriminant Analysis": {},  # No hyperparameters to tune
}

# Initialize classifiers
classifiers = {
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting Machine": GradientBoostingClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Gaussian Naive Bayes": GaussianNB(),
    "Linear Discriminant Analysis": LinearDiscriminantAnalysis(),
}

# Store evaluation metrics
results = []

# Train and evaluate models with hyperparameter tuning
for name, clf in classifiers.items():
    print(f"Training {name}...")

    # Retrieve hyperparameter grid for the model
    param_grid = param_grids.get(name, {})

    # Use GridSearchCV for hyperparameter tuning
    grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Get the best model
    best_clf = grid_search.best_estimator_

    # Predictions
    y_pred = best_clf.predict(X_test)
    y_prob = best_clf.predict_proba(X_test)[:, 1] if hasattr(best_clf, "predict_proba") else None

    # Generate and display the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])  # Assuming binary classification
    disp.plot(cmap='Blues', values_format='d')
    plt.title(f"Confusion Matrix for {name}")
    plt.show()

    # Evaluation metrics
    metrics = {
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-Score": f1_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_prob) if y_prob is not None else "N/A",
        "Best Params": grid_search.best_params_,
    }
    results.append(metrics)

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
```

```python
# Display results in a nice table using tabulate
from tabulate import tabulate
print("\nEvaluation Results:")
print(tabulate(results_df, headers='keys', tablefmt='fancy_grid', showindex=False))
```
